```python
matchtype = pd.get_dummies(df.matchType,dtype=int,prefix='matchType')
matchtype.head()
```

```python
pd.concat([df,matchtype],axis=1)
```

```python
df.drop("matchType",axis=1,inplace=True)
```

```python
df.columns
```

```python
for i in df.iloc[:,3:]:
  q1 = df[i].quantile(0.25)
  q3 = df[i].quantile(0.75)
  iqr = q3-q1
  lower = q1 - 1.5*iqr
  upper = q3 + 1.5*iqr
  df.loc[(df[i]<lower)|(df[i]>upper),i] = df[i].median()
```

```python
x = df.drop(['winPlacePerc','Id','groupId','matchId'],axis=1)
y = df['winPlacePerc']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=33)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
r2_score(y_test,y_pred)
```
